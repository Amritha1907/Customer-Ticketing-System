##Add More Tickets and collect details

```python
No_of_Ticket= int(input("How many new tickets do you want to add?""\n"))
ii=10
for i in range(No_of_Ticket):
  Customer_Name=input("Enter Customer Name: ")
  Issue_Description= input("Enter Issue Description: ")
  Priority=input("Enter Priority - High/ Medium/ Low: ").lower()

  while Priority not in ["high","low","medium"]:
    print("Invalid Priority")
    Priority=input("Enter Priority - High/ Medium/ Low: ").lower()
  ii+=1
  ticket_data['Ticket_No'].append(ii)
  ticket_data['Customer_Name'].append(Customer_Name)
  ticket_data['Issue_Description'].append(Issue_Description)
  ticket_data['Priority'].append(Priority)
  print("-"* 30)
print(ticket_data)
```

##3. Text Cleaning for Issue Descriptions

```python
for i in range(len(ticket_data['Issue_Description'])):
  text = ticket_data['Issue_Description'][i]
#Remove punctuation (.,!?-)
  text = text.replace(".", "").replace(",", "").replace("!", "").replace("?", "").replace("-", "")
#Convert multiple spaces → single space
  text=text.strip()
#Remove leading/trailing spaces
  text=' '.join(text.split())
#Convert text to lowercase
  text=text.lower()
#Replace slang/shorthand (example: "ok" → "okay")
  text = text.replace("ok", "okay").replace("hi", "hai").replace("hlo", "hello").replace("nah", "no").replace("yeah", "yes")
  ticket_data['Issue_Description'][i]=text
print(ticket_data['Issue_Description'])
```

##4. Count the mentioned words using keyword function in Issue Description

```python
def count_tickets_with_word(word):
  count=0
  for i in ticket_data['Issue_Description']:
    if word in i:
      count+=1
  return count
print("Number of tickets containing - poor", count_tickets_with_word('poor'))
print("Number of tickets containing - good", count_tickets_with_word('good'))
print("Number of tickets containing - slow", count_tickets_with_word('slow'))
print("Number of tickets containing - excellent", count_tickets_with_word('excellent'))
```
